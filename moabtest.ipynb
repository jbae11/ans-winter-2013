```python
import yt
print yt.__version__
from yt.mods import *
import tables as tb
import numpy as np
```

```python
f = tb.open_file('flux_mesh_simple.h5m', 'r')
```

```python
# raw data
con = f.root.tstt.elements.Hex8.connectivity[:]
coord = f.root.tstt.nodes.coordinates[:]
n_group_001 = f.root.tstt.elements.Hex8.tags.n_group_001[:]
n_group_001_error = f.root.tstt.elements.Hex8.tags.n_group_001_error[:]
```

```python
domdims = coord[[0,-1]].T
cell_centers = coord[con-1].sum(axis=1)/8
uvalues = map(lambda x: np.unique(cell_centers[:, x]), range(3))
idx = np.empty(map(len, uvalues), int)
```

```python
for xi, xcoord in enumerate(uvalues[0]):
    for yi, ycoord in enumerate(uvalues[1]):
        for zi, zcoord in enumerate(uvalues[2]):
            idx[xi,yi,zi] = np.argwhere(np.all(cell_centers == (xcoord, ycoord, zcoord), axis=1))[0][0]
```

```python
fields = {'n_group_001': n_group_001[idx], 'n_group_001_error': n_group_001_error[idx]}
```

```python
pf = load_uniform_grid(fields, idx.shape, 1, bbox=domdims)
s = SlicePlot(pf, 'z', 'n_group_001', origin='native')
s.set_log('n_group_001', True)
linestyle = {'color': 'k'}
s.annotate_image_line([15, 0.1], [15, 65], True, linestyle)
s.annotate_image_line([10, 0.1], [10, 60], True, linestyle)
s.annotate_image_line([-15, 0.1], [-15, 65], True, linestyle)
s.annotate_image_line([-10, 0.1], [-10, 60], True, linestyle)
s.annotate_image_line([-15, 65], [15, 65], True, linestyle)
s.annotate_image_line([-15, 60], [15, 60], True, linestyle)
s.annotate_image_line([5, 0.1], [5, 55], True, linestyle)
s.annotate_image_line([-5, 0.1], [-5, 55], True, linestyle)
s.annotate_image_line([-5, 55], [5, 55], True, linestyle)
fig = s.plots[s.fields[0]]
fig.cb.set_label("Neutron Flux in Group 001 ($n/cm^2$)")
s.display()
```
