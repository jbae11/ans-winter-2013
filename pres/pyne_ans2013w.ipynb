## What is PyNE?

As out tagline says, PyNE is the open source nuclear engineering toolkit.  

* PyNE is intended as a library of composable tools that is used to build nuclear science and engineering applications.
* It is permissively licensed (2-clause BSD)
* It supports both a C++ and a Python API.
* The name 'PyNE' is a bit of a misnomer since most of the code base is in C++ but most daily usage happens in Python.
* The v0.3 code base was just released!
* As an organization, PyNE was born in April 2011.  However, core parts of PyNE have existed since 2007.

## What are the goals of PyNE?

To help nuclear engineers:

* be more productive,
* have the *best* solvers,
* have a beautiful API,
* write really great code,
* and teach the next generation.

And of course, to perform neat tricks like this:

````python
import numpy as np
from pyne import mesh, material 
from pyne.xs.channels import sigma_t
from yt.config import ytcfg; ytcfg["yt","suppressStreamLogging"] = "True"
from yt.mods import *
m = mesh.Mesh(structured_coords=[np.linspace(0.0, 1.0, 101), np.linspace(0.0, 1.0, 101), [0.0, 1.0]], structured=True)
fuel = material.from_atom_frac({'U235': 0.045, 'U238': 0.955, 'O16': 2.0}, mass=1.0, density=10.7)
cool = material.from_atom_frac({'H1': 2.0, 'O16': 1.0}, mass=1.0, density=1.0)
for i, mat, ve in m:
    coord = m.mesh.getVtxCoords(ve)
    m.mats[i] = fuel if (coord[0]**2 + coord[1]**2) <= 0.5**2 else cool
m.sigma_t = mesh.ComputedTag(lambda mesh, i: sigma_t(mesh.mats[i], group_struct=[10.0, 1e-6], phi_g=[1.0])[0])
m.u_mass = mesh.ComputedTag(lambda mesh, i: max(mesh.mats[i]['U':'Np'].mass, 0.0))
s = SlicePlot(PyneMoabHex8StaticOutput(m), 'z', 'sigma_t', origin='native')
````

````python
print "Pu Content of Mesh:", sum(m.u_mass[:])
s.display()
````

## endftod()

Floating point ENDF data has a particular format:

<div style="text-align:center;"><img src="endf_nums.svg" width="80%" /></div>

Converting the string representation of these numbers to floats in memory by adding an `'e'` is expensive for two reasons:

1. Adding the exponent requires string manipulations
2. The `atof()` and `atod()` calls are expensive!

<div style="text-align:right;"><h3>We can do better!</h3></div>

## endftod()

Because this is a fixed-width format with sematic meaning on placement, there are a number of optimizations possible.

The naive Python implementation is:

```python
float(x[:-3] + 'e' + x[-3] + x[-2:])
```

````python
t = [['Implementation', 'Average Run Time [ns]', 'Speed Up'], 
     ['Naive Python, float()', 539,  1.0],
     ['Naive C, atof()', 229, 2.3],
     ['PyNE, endftod()', 114, 4.7],
     ]
x = htmltable(t, 'endftod', title="ENDF Float Conversion Comparison", style={'text-align': 'center'})
print x[0]
print x[1]
````

## Meshes

PyNE also has a mesh interface that is used to represent any kind of mesh desired.  This is a layer on top of [MOAB](https://trac.mcs.anl.gov/projects/ITAPS/wiki/MOAB) meshes which adds:

1. PyNE materials to volume elements
2. A generic tagging interface.

In the `pyne.mesh` module lives the `Mesh` class.  This class houses an iMesh instance and contains methods for various mesh operations. The following attribute of a `Mesh` object are most important.

* **mesh:** iMesh instance

* **mats:** MaterialLibrary or dict or Materials or None, This is a mapping of volume element handles to Material objects.

To do any visulaization or volumetric analysis of such meshes, we recomend the [yt project](http://yt-project.org/).

````python
def cantor(n):
    return [0.] + cant(0., 1., n) + [1.]

def cant(x, y, n):
    if n == 0:
        return []
    new_pts = [2.*x/3. + y/3., x/3. + 2.*y/3.]
    return cant(x, new_pts[0], n-1) + new_pts + cant(new_pts[1], y, n-1)
````

````python
c5 = cantor(5)
coords = [c5, c5, [0.0, 1.0]]
````

## Sample Mesh Setup

````python
from pyne.mesh import Mesh, IMeshTag, MetadataTag, ComputedTag
m = Mesh(structured_coords=coords, structured=True)
````

Put materials on the mesh:

````python
from pyne.material import Material, from_atom_frac
fuel = from_atom_frac({'U235': 0.045, 'U238': 0.955, 'O16': 2.0}, density=10.7)
cool = from_atom_frac({'H1': 2.0, 'O16': 1.0}, density=1.0)
for i in range(len(m)):
    m.mats[i] = cool
m.mats[len(m)/2] = fuel
m.mats[len(m)/4] = fuel
for i, c in enumerate(c5[:-1]):
    m.mats[i*len(c5)] = fuel
````

## Mesh Tags

`Tag`s - sometimes known as fields - are generic way of storing data on a mesh.  Tags can be accessed as attributes on the mesh class itself.  For example, the density tag may be grabbed via:

````python
m.density 
````

To get the value of the density, you have to provide the location by its volume element index, slice, mask, or fancy index:

````python
print "index:", m.density[42]
print "slice:", m.density[::100]
print "mask:", m.density[m.density[:] >= 10]
print "fancy:", m.density[[10, 0, 11, 100]]
````

## Computed Tags

Computed tags - also known as 'derived fields' - are a way of having a lazily evaluated 'virtual tag.'  The following example which uses PyNE's cross section tools to compute the one-group total cross section $\sigma_t$ everywhere on the mesh.

````python
from pyne.xs.channels import sigma_t
m.sigma_t = ComputedTag(lambda mesh, i: sigma_t(mesh.mats[i], group_struct=[10.0, 1e-6], phi_g=[1.0])[0])
s = SlicePlot(PyneMoabHex8StaticOutput(m), 'z', 'sigma_t', origin='native').display()
````
